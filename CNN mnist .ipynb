```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Dropout
from tensorflow.keras.utils import to_categorical
```

```python
df=pd.read_csv('mnist_train.csv')
```

```python
df_test=pd.read_csv('mnist_test.csv')
```

```python
# preprocessing 
```

```python
X_train=df.drop('label',axis=1)
y_train=df['label']
```

```python
X_test=df_test.drop('label',axis=1)
y_test=df_test['label']
```

```python
# there are values netween 0 to 255    convert thoose values between 0 and 1 
```

```python
X_train=X_train.astype('float32')/255
```

```python

X_test=X_test.astype('float32')/255
```

```python

print(X_train.shape)
print(X_test.shape)
```

```python
# reshape the data 
```

```python
X_train_img = X_train.values.reshape(-1, 28, 28)
X_test_img = X_test.values.reshape(-1, 28, 28)
```

```python
# Each row of 784 numbers is rearranged into a 28Г—28 image. 
# in the place of -1 u can use 60000 or 10000 as well 
```

```python
df['label'].unique()
```

```python
y_test_cat=to_categorical(y_test,10)  # total 10  labels are presented in the label thast why 
```

```python
y_train_cat=to_categorical(y_train,10)
```

```python
perceptron=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(10,activation='softmax')])
```

```python
  # here we created two layers 
# first layer by flatten as 785 inputs in the shape of 28*28 
# second as output layer for 20 labels and activation function of softmax 
```

```python
perceptron.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=perceptron.fit(X_train_img,y_train_cat,batch_size=30,epochs=10,verbose=1,validation_data=[X_test_img,y_test_cat])
```

# Why do we use validation?

# During training, the model learns from the training data. If you only look at training accuracy, the model might simply memorize the training set (overfitting).

```python
loss,acc=perceptron.evaluate(X_test_img,y_test_cat,verbose=1)
```

```python
# now ann using 
```

```python
model=Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation='relu'),
    Dense(64,activation='relu'),
    Dense(32,activation='relu'),
    Dense(10,activation='softmax')])
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
history1=model.fit(X_train_img,y_train_cat,batch_size=32,epochs=10,verbose=1,validation_data=[X_test_img,y_test_cat])
```

```python
acc=model.evaluate(X_test_img,y_test_cat,verbose=1)
```

```python
# CNN 
```

```python
X_train_cnn=X_train.values.reshape(-1,28,28,1)
X_test_cnn=X_test.values.reshape(-1,28,28,1)
```

```python
# here the last 1 is callles channel colour 
# if it is 1  used for  black and white 
# if it is 3 then used for red green blue (RGB)
```

```python
cnn=Sequential([
    Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64,kernel_size=(3,3),activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(120,activation='relu'),
    Dropout(0.5),                      # to protect from overfitting 
    Dense(10,activation='softmax')])
    
```

```python
cnn.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=cnn.fit(X_train_cnn,y_train_cat,batch_size=30,epochs=10,verbose=1,validation_data=[X_test_cnn,y_test_cat])
```

```python
acc=cnn.evaluate(X_test_cnn,y_test_cat,verbose=1)
```
